## Key Themes
- No Big Bang Integration — add one module at a time
- Skeleton top module with heartbeat → synthesize first
- UART printf debugging for hardware-only issues
- Minimum viable demo: something must work on hardware by Day 16
- Individual check-ins: status, block diagram review, scope management
