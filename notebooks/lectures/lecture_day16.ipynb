## Demo Format (5–7 min per student)

1. Introduction (30s): project name, one sentence
2. Live demo (2m): show it working on Go Board
3. Architecture (1–2m): block diagram walk-through
4. Verification (1m): testbench results, assertions
5. Reflection (1m): challenges, what you'd change
6. Q&A (remaining)

## Evaluation Weights

- Functionality: 30%
- Design Quality: 25%
- Verification: 20%
- Integration: 15%
- Presentation: 10%
